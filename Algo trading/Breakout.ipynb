# Breakouts

In technical analysis, one way to trade is when the breakout occurs. Breakout can occur when a price suddenly jumps after a period of moving sideways or in the same direction in the market. This jump is usually the result of an accumulation phase. 

When the price is moving sideways, it tends to accumulate a lot of positions. Once there are enough positions on one side (buying or selling), the pressure builds up. And it suddenly bursts, resulting in a jump in the price. You can say that the price breakout is due to a build-up of momentum.

Similarly, when the price is moving in the same direction for a given period, it tends to lose momentum after a while. There is an accumulation of positions on the other side, leading to a reversal and a breakout.

---

But why do we use the concept of a breakout?

Asset prices are not trending most of the time and tend to be in a sideways market. In a sideways market, a breakout strategy helps us maximise our gains. Also, a breakout can signal the start of a trend and thus, as early movers, you will have an advantage.

## Implementation of Breakout Indicator  

Let's use the same example of stock Tesla from the previous notebook. And plot its closing price along with the curve portraying the maximum closing price of Tesla in the last 5 days.

```python
# For data manipulation
import pandas as pd

# Import matplotlib as an alias plt and set the style
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

# Read Tesla price data from a csv file
tesla_price = pd.read_csv('../data/tesla_stock_price.csv', index_col=0)

# Convert index to datetime format
tesla_price.index = pd.to_datetime(tesla_price.index)

# Calculate rolling maximum close price
tesla_price['max'] = tesla_price.Close.rolling(window=5).max()

# Plot close price and rolling maximum close price
tesla_price[['Close','max']].plot(figsize=(15,7))
plt.title('Maximum Closing Price of Stock Tesla', fontsize=14)
plt.ylabel('Price',fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.show()
```

Here, you can interpret the above chart as follows,

* If the line of the maximum price is at the same level as Teslas closing price level, then the closing price is increasing. This could imply a bull trend. 
* If the closing price dips below the maximum price line, it could signal a bear trend.

Thus, in this way, we have tried to interpret the breakout signals and identify the trends.<br><br>
