```python
import json
import requests
from pandas_datareader import data as web
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# Получение данных по опционам из АПИ Эксанты. Данные получаются в Json массиве.
response= requests.get("https://api-live.exante.eu/md/3.0/types/OPTION/", auth=("1beca0f1-7643-49ba-9643-eed5905fe837", "a8XQAmy9eA7N80soYAZS")).json()
response
```

```python
# Перевод Json в Pandas DataFrame. Но эта функция делает перевод довольно коряво, 
# т.к. в изначальном массиве есть разная вложенность элементов
df = pd.json_normalize(response)
```

```python
df.sort_values(by='ticker')
myset = set(df['ticker'])
```

```python
# Получение столбцов Биржи и Тикера
df1 = df['exchange'].apply(lambda row: (row))
df2 = df['ticker'].apply(lambda row: (row))
```

```python
# Подготовка полученных данных и объединение в один дата фрейм. 
# Теперь к данным можно обращаться по названию столбца
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
df3 = df1.join(df2, how='outer', rsuffix='_')
```

```python
# Фильтрация всех тикеров на чикагской бирже CBOE. И фильтрация уникальных тикеров на этой бирже. 
# Получили 955 тикеров
df3 = pd.DataFrame(df3)
df4 = df3.loc[df3['exchange']=='CBOE']
df5 = df4.drop_duplicates('ticker')
df5
```

```python
# Запись итоговой таблицы в файл
import csv 
df5.to_csv("options-on-CBOE.csv", sep='\t')
```
